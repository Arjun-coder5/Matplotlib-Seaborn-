# ***Distribution Plot***

````python
import seaborn as sns
import matplotlib.pyplot as plt
````



```
# This is formatted as code
We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

distplot/histplot
jointplot
pairplot
rugplot
```



````python
df = sns.load_dataset('tips')
df
````

````python
df['size'].unique()
````

````python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
````

````python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
````

````python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'kde')
````

````python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'hex')
````

````python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'reg')
````

````python
sns.pairplot(df,hue = 'sex')
````

````python
sns.pairplot(df,hue = 'size',palette = 'rainbow')
````

````python
sns.rugplot(df['total_bill'])
````

# ***Matrix*** Plot :

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
flights = sns.load_dataset('flights')
flights
````

````python
tips = sns.load_dataset("tips")
tips
````

````python
tipscorr  = tips[['total_bill','tip','size']]
````

````python
tipscorr.corr()
````

````python
sns.heatmap(tipscorr.corr(),annot= True)
````

````python
sns.clustermap(tipscorr.corr())
````

````python
pvtflight = flights.pivot_table(values = 'passengers',index = 'month',columns = 'year')
````

````python
pvtflight
````

````python
sns.heatmap(pvtflight)
````
